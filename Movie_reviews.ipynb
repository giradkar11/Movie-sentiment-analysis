```python
neg_files_path=folder + 'neg'
pos_files_path= folder +'pos'

neg_text_file=glob.glob(f"{neg_files_path}\\*.txt")
pos_text_file=glob.glob(f"{pos_files_path}\\*.txt")
```

```python
review_list=[]
for file_path in neg_text_file:
    f=open(file_path)
    txtfile=f.read()
    txtfile=re.sub('[^a-zA-Z]+',' ',txtfile)
    f.close()
    review_list.append(txtfile)
    
for file_path in pos_text_file:
    f=open(file_path)
    txtfile=f.read()
    txtfile=re.sub('[^a-zA-Z]+',' ',txtfile)
    f.close()
    review_list.append(txtfile)
    
print(len(review_list))
```

```python
we have 2000 row now 
(x)feature to be created
```

```python
cntv=CountVectorizer(stop_words='english', min_df= 0.05)
x_cntv=cntv.fit_transform(review_list)

x=pd.DataFrame(x_cntv.toarray(),columns=cntv.get_feature_names_out())
x
```

```python
tfidV=TfidfVectorizer(stop_words='english' , min_df= 0.05)
x_tfidV=tfidV.fit_transform(review_list)

x=pd.DataFrame(x_tfidV.toarray(),columns=tfidV.get_feature_names_out())
x
```

```python
create target columns 
nagative reviw =0
positive review =1
```

```python
positive_target=np.ones(1000,dtype='int')
negative_target=np.zeros(1000,dtype='int')

y=np.append(negative_target,positive_target)
y=pd.Series(y)
y
```

```python
model building
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1,stratify=y)
```

# GaussianNB

```python
Gnb=GaussianNB()
Gnb.fit(x_train,y_train)
```

```python
# Evaluation training
y_pred_train = Gnb.predict(x_train)

print('classification report :', classification_report(y_pred_train,y_train))

print('accuracy score :', accuracy_score(y_pred_train,y_train))
```

```python
# Evaluation testing
y_pred_test = Gnb.predict(x_test)

print('classification report :', classification_report(y_test,y_pred_test))

print('accuracy score :', accuracy_score(y_test,y_pred_test))
```

# MultinomialNB

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
# Evaluation training
y_pred_train = mnb.predict(x_train)

print('classification report :', classification_report(y_pred_train,y_train))

print('accuracy score :', accuracy_score(y_pred_train,y_train))
```

```python
# Evaluation testing
y_pred_test = mnb.predict(x_test)

print('classification report :', classification_report(y_test,y_pred_test))

print('accuracy score :', accuracy_score(y_test,y_pred_test))
```

# BernoulliNB

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
# Evaluation training
y_pred_train = bnb.predict(x_train)

print('classification report :', classification_report(y_pred_train,y_train))

print('accuracy score :', accuracy_score(y_pred_train,y_train))
```

```python
# Evaluation testing
y_pred_test = bnb.predict(x_test)

print('classification report :', classification_report(y_test,y_pred_test))

print('accuracy score :', accuracy_score(y_test,y_pred_test))
```
